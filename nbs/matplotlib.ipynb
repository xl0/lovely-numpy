````python
# |hide
numbers = np.load("mysteryman.npy").transpose(1,2,0)
in_stats = ( (0.485, 0.456, 0.406),     # mean
             (0.229, 0.224, 0.225) )    # std

mean = np.array(in_stats[0])
std = np.array(in_stats[1])
numbers = (numbers*std + mean).clip(0,1)
````

### <code>.fig</code>

`.rgb`, `.chans` and `.plt` all have a `.fig` attribute that returns a matplotlib figure object. 

````python
a = lo(numbers).rgb.fig # matplotlib figure
print(type(a))
a
````

````python
lo(numbers).chans.fig
````

````python
lo(numbers).plt.fig
````

````python
lo(numbers).plt(center="mean").fig
````

### Using existing `Axes`

All functions allow an `ax=` argument that accepts an existing `Axes` object into which they will plot:

````python
fig = plt.figure(figsize=(8,3))
fig.set_constrained_layout(True)
gs = fig.add_gridspec(2,2)
ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1, 0])
ax3 = fig.add_subplot(gs[1,1:])

ax2.set_axis_off()
ax3.set_axis_off()

lo(numbers).plt(ax=ax1)
lo(numbers).rgb(ax=ax2)
lo(numbers).chans(ax=ax3);
````

### Without Jupyter

By default, the Lovely functions will call `plt.close(fig)` on the figures they create.

This prevents displaying the figures twice when running in Jupyter.\
If you are not using Jupyter, here are 2 [configuration options](https://xl0.github.io/lovely-numpy/03d_utils.config.html) you might want to set:

`fig_close=False`

```py
#!/usr/bin/env python
from lovely_numpy import config, set_config, lo

...

set_config(fig_close=False)
lo(numbers).chans()

# or, using the context manager:
with config(fig_close=False):
    lo(numbers).chans()

plt.show() # Will show all open figures
```

`fig_show=True`

If set, lovely will call `plt.show()` after each figure creation.\
You don't need to set `fig_close=False` manually.

```py
set_config(fig_show=True)

lo(numbers).chans() # Figure generated and shown

# Note, you have to use the "call" syntax `( )`, as figure
# generation is not triggerd by just accessing the attribute

lo(numbers).chans  # No figure generated

f = lo(numbers).plt.fig # figure generated, shown, and returned.
```

Note, `plt.show()` closes all figures.
